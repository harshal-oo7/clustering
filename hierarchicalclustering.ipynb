```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("/Users/harshalghodmare/Desktop/shopping_data.csv")
df.head()
```

```python
data=df.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
plt.title("customer dendogram")
dendo=sch.dendrogram(sch.linkage(data ,method='single'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_)
```
